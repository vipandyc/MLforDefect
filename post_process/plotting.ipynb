```python
import numpy as np
import matplotlib.pyplot as plt
import re
from tqdm import tqdm
plt.rcParams.update({'figure.max_open_warning': 0})

# Function to read a specific line from a file
def read_line_from_file(filename, line_number):
    with open(filename, 'r') as file:
        lines = file.readlines()
        if line_number < len(lines):
            return lines[line_number].strip()
        else:
            raise ValueError(f"Line {line_number} does not exist in {filename}")


fine_bins = np.linspace(0, 0.05, num=50)  # 50 bins between 0 and 0.05 for fine sampling
coarse_bins = np.linspace(0.05, 0.25, num=10)  # 10 bins for coarser sampling

# Combine the fine and coarse bins, ensuring no duplicate at 0.05
bin_edges = np.unique(np.concatenate((fine_bins, coarse_bins)))


'''
for i in tqdm(range(468)):
    # Read the ith line from eigenvalues_list.txt and convert to floats
    eigenvalues_line = read_line_from_file('eigenvalues_list.txt', i)
    eigenvalues = list(map(float, eigenvalues_line.split()))

    # Read the ith line from poscar_elements_list.txt for the title
    poscar_elements_line = read_line_from_file('poscar_elements_list.txt', i)

    # Format the title to insert spaces between elements and numbers
    formatted_title = ' '.join(re.findall(r'[A-Za-z]+|\d+', poscar_elements_line))

    # Plot the histogram for the density of states (DoS)
    plt.figure(figsize=(4, 3))
    plt.hist(eigenvalues, bins=bin_edges, density=True, alpha=0.7, color='blue')
    plt.title(f"Index {i}: {formatted_title}")
    plt.xlabel("Energy(eV)")
    plt.ylabel("DoS")
    plt.grid(True)
    plt.tight_layout()

    # Display the plot
    plt.savefig(f'datafigs/Fig {i}.jpg')
    plt.clf()'''
```

```python
plt.figure(figsize=(6, 4))
color = ['blue', 'red']
label = ['5% P doping', '5% B doping']
for j, i in enumerate([345, 7]):
    # Read the ith line from eigenvalues_list.txt and convert to floats
    eigenvalues_line = read_line_from_file('eigenvalues_list.txt', i)
    eigenvalues = list(map(float, eigenvalues_line.split()))

    # Read the ith line from poscar_elements_list.txt for the title
    poscar_elements_line = read_line_from_file('poscar_elements_list.txt', i)

    # Format the title to insert spaces between elements and numbers
    formatted_title = ' '.join(re.findall(r'[A-Za-z]+|\d+', poscar_elements_line))

    # Plot the histogram for the density of states (DoS)
    
    plt.hist(eigenvalues, bins=bin_edges, density=True, alpha=0.7, color=color[j], label=label[j])
    #plt.title(f"Index {i}: {formatted_title}")
    plt.xlabel("Energy(eV)")
    plt.ylabel("DoS")
    plt.grid(True)
    plt.tight_layout()

    # Display the plot
plt.legend()
plt.show()
```
