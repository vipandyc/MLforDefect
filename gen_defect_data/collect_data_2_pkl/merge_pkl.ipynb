```python
import pandas as pd
from tqdm import tqdm
# Load the two CSV files
csv1_path = "eigvals_data_2025-02-26_chk1.pkl"  # Replace with actual file path
csv2_path = "eigvals_data_2025-02-26_chk2.pkl"  # Replace with actual file path

df1 = pd.read_pickle(csv1_path)
df2 = pd.read_pickle(csv2_path)

# Merge the dataframes
merged_df = pd.concat([df1, df2], ignore_index=True)

# Create an empty DataFrame for storing the processed data
processed_data = []

for _, entry in tqdm(merged_df.iterrows(), total=len(merged_df)):
    if entry["j"] > 0:
        j0_entry = merged_df[(merged_df["i"] == entry["i"]) & (merged_df["j"] == 0)]
        if not j0_entry.empty:
            processed_data.append({
                "formula_j0": j0_entry.iloc[0]["formula"],
                "formula_j": entry["formula"],
                "eigvals_j0": j0_entry.iloc[0]["eigvals"],
                "eigvals_j": entry["eigvals"],
                "dopant_options_j0": j0_entry.iloc[0]["dopant_options"]
            })
        else:
            print(f'Error in formula {entry["formula"]}')

processed_df = pd.DataFrame(processed_data)
processed_df.to_pickle('../../DefectNet/DefectNet_data_v0.1.pkl')
processed_df.head(30)
```
